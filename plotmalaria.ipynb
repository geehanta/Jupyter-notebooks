```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load your dataset
df = pd.read_csv('bbpsymptoms.csv')
```

```python
# Assuming 'pathogen' is the column containing the pathogen names
# Assuming 'malaria' is one of the pathogens you want to analyze
malaria_data = df[df['Pathogen'] == 'Malaria']
```

```python
# Select only the symptom columns
symptoms = df.columns[0:]  # Assuming the symptoms start from the second column
```

```python
plt.figure(figsize=(10, 8))
symptom_corr = malaria_data[symptoms].corr()
sns.heatmap(symptom_corr, annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap for Malaria Symptoms')
plt.show()
```
